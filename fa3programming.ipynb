```python
#fa3prog3
name = input("Enter your name: ")
score = float(input(f"Enter exam score for {name}: "))

if score >=90 and score <= 100:
  grade = "A"
elif score >=80:
  grade = "B"
elif score >=70:
  grade = "C"
else:
  grade = "F"

print(f"Student: {name}")
print(f"Score: {score}")
print(f"Grade: {grade}")
```
